### Importing libraries

```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

import server as srv
import agent as agn
```

Setting a random seed

```python
np.random.seed(0)
```

### Defining the number of ports

```python
nports = 16
```

## Simulation 2.1

We run a limited number of episodes ($100$) with *probdetection*=0.0. 

```python
probdetection = 0.0
nepisodes = 100
```

```python
agt = agn.Agent(nports)
agt.set_learning_options(exploration=0.2, 
                         learningrate=0.1, 
                         discount=0.9, 
                         scanprobability=0.1)
```

```python
steps = []
for _ in range(nepisodes):
    
    vulnport = np.random.randint(1,nports+1)
    env = srv.Server(nports,vulnport,probdetection)
    agt.reset(env)
    
    agt.run_episode()
    steps.append(agt.steps)
```

```python
plt.matshow(agt.Q)
```

As before, we mostly learn a diagonal Q matrix.

```python
plt.plot(np.arange(nepisodes),steps)
```

We also notice a clear decrease in the number of steps per episode as the agent learns an optimal strategy (remember though that the agent always acts using an $\epsilon$-strategy).

## Simulation 2.2
We now consider a highly sensitive server able to easily detect scan attempts. We set *probdetection*=0.8.

```python
probdetection = 0.8
nepisodes = 100
```

```python
agt = agn.Agent(nports)
agt.set_learning_options(exploration=0.2, 
                         learningrate=0.1, 
                         discount=0.9, 
                         scanprobability=0.1)
```

```python
steps = []
for _ in range(nepisodes):
    
    vulnport = np.random.randint(1,nports+1)
    env = srv.Server(nports,vulnport,probdetection)
    agt.reset(env)
    
    agt.run_episode()
    steps.append(agt.steps)
```

```python
plt.matshow(agt.Q)
```

We now lose the diagonal pattern as, very likely, after scanning the flag will be moved. Here random actions, especially at the very beginning, pay off the best.

```python
plt.plot(np.arange(nepisodes),steps)
```

We do not notice any decrease in the average number of step, as the game with such a *probdetection* renders the scan action useless and prevents the development of a strategy.

## Simulation 2.5
We run simulations with different values of probdetection for $100$ times in order to collect statistics.

```python
ntrials = 100

probdetections = np.linspace(0,1,11)
nprobdetection = len(probdetections)

nepisodes = 1000

steps = np.zeros((ntrials,nprobdetection,nepisodes))
Q_diag = np.zeros((ntrials,nprobdetection,nepisodes))
Q_ndiag = np.zeros((ntrials,nprobdetection,nepisodes))
Qs = np.zeros((ntrials,nprobdetection,nports+1,nports+1))

for t in tqdm(range(ntrials)):

    agt = agn.Agent(nports,verbose=False)
    agt.set_learning_options(exploration=0.2, 
                             learningrate=0.1, 
                             discount=0.9, 
                             scanprobability=0.1)

    for p in range(nprobdetection):
        probdetection = probdetections[p]
    
        nepisodes = 1000
        for e in range(nepisodes):

            vulnport = np.random.randint(1,nports+1)
            env = srv.Server(nports,vulnport,probdetection)
            agt.reset(env)

            agt.run_episode()

            steps[t,p,e] = agt.steps
            Q_diag[t,p,e] = np.sum(np.diag(agt.Q))
            Q_ndiag[t,p,e] = np.sum(agt.Q)
            Qs[t,p,:,:] = agt.Q
```

```python
plt.matshow(np.mean(Qs[:,1,:,:],axis=0))

plt.xlabel('Actions')
plt.ylabel('States')
plt.tick_params(axis='x',bottom=True,top=False,labelbottom=True,labeltop=False)

clb = plt.colorbar()
clb.ax.set_title('Q-value')

plt.savefig('simul2_matrix1')
```

```python
plt.matshow(np.mean(Qs[:,5,:,:],axis=0))

plt.xlabel('Actions')
plt.ylabel('States')
plt.tick_params(axis='x',bottom=True,top=False,labelbottom=True,labeltop=False)

clb = plt.colorbar()
clb.ax.set_title('Q-value')

plt.savefig('simul2_matrix2')
```

```python
plt.matshow(np.mean(Qs[:,10,:,:],axis=0))

plt.xlabel('Actions')
plt.ylabel('States')
plt.tick_params(axis='x',bottom=True,top=False,labelbottom=True,labeltop=False)

clb = plt.colorbar()
clb.ax.set_title('Q-value')

plt.savefig('simul2_matrix3')
```

```python
plt.plot(range(nepisodes),np.mean(steps[:,10,:],axis=0))
```

```python
plt.plot(range(nepisodes),np.mean(steps[:,1,:],axis=0),label='p=0.1',linestyle='dotted')
plt.plot(range(nepisodes),np.mean(steps[:,5,:],axis=0),label='p=0.5',linestyle='dashed')
plt.plot(range(nepisodes),np.mean(steps[:,10,:],axis=0),label='p=1.0')

plt.xlabel('Episodes')
plt.ylabel('Number of steps')
plt.legend()
plt.grid()

plt.savefig('simul2_steps')
```
